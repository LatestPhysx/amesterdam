```python
import pandas as pd
import numpy as np
from pathlib import Path
# dates & cleaning
import re
import warnings
warnings.filterwarnings("ignore")
# optional: show full tables in console
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 120)
```

```python
students_file = "../DATA/Students per education type Amsterdam.csv"
```

```python
students = pd.read_csv(students_file, sep=";")
```

```python
students.head()
```

```python
years = students.iloc[0, 2:]
students_clean = students.iloc[2:].copy()
```

```python
students_clean.columns = ["education_type", "age_group"] + list(years)
```

```python
students_clean.head(7)
```

```python
students_clean = students_clean.drop(columns="age_group")
```

```python
students_long = students_clean.melt(
    id_vars="education_type",
    var_name="year",
    value_name="students"
)
```

```python
students_long["students"] = students_long["students"].astype(int)
```

```python
students_long["students"] = pd.to_numeric(students_long["students"], errors="coerce")
students_long = students_long.dropna(subset=["students"])
students_long["students"] = students_long["students"].astype(int)
```

```python
students_long["year"] = (
    students_long["year"]
    .astype(str)
    .str.replace(r"\D", "", regex=True)   # keep only digits
    .str[-2:]                             # last two digits
    .astype(int) + 2000
)
```

```python
students_long.to_csv("../data_clean/students.csv", index=False)
```
